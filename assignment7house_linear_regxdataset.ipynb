```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv(open("D:/ds/datasets/house_linear_regx.csv"))
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.sort_values(by=['date'])
df
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(ax=ax, data=df.corr(), cmap='coolwarm', annot=True)
```

```python
df1 = df.drop(columns=['price', 'id'])
df1
```

```python
x=df1
```

```python
y = df['price']
y
```

```python
x=np.array(x)
x
y=np.array(y)
y
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
x[:,0]=lbl.fit_transform(x[:,0])
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
ip=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(ip,y,test_size=0.2) 
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)  
predlr=lr.predict(xtest) 
```

```python
print(xtrain.shape)
print(ytrain.shape)
```

```python
predlr1 = predlr[0:100]
ytest1 = ytest[0:100]
xaxis = np.linspace(1,len(predlr1), len(predlr1))
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(xaxis, predlr1, color='red')
plt.plot(xaxis, ytest1, color='blue')
plt.show()
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(ytest1, predlr1))
print('MSE:', metrics.mean_squared_error(ytest1, predlr1))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest1, predlr1)))
```

```python
sns.jointplot(x=ytest, y=predlr, kind='reg', joint_kws={'line_kws':{'color':'red'}});
```

```python
from sklearn.ensemble import RandomForestRegressor
rr = RandomForestRegressor()
rr.fit(xtrain,ytrain)  
predrr=rr.predict(xtest)
predrr1 = predrr[0:100]
ytest2 = ytest[0:100]
xaxis2 = np.linspace(1,len(predrr1), len(predrr1))
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(xaxis2, predrr1, color='red')
plt.plot(xaxis2, ytest2, color='blue')
plt.show()
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(ytest2, predrr1))
print('MSE:', metrics.mean_squared_error(ytest2, predrr1))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest2, predrr1)))
```
